## ‚≠ê Linear Regression 

Linear Regression is a statistical and machine-learning method used to **predict one or more variables** based on **one or more other variables**.

- The variable we want to predict is called the **dependent variable (y)**.
- The variables used to make the prediction are called **independent variables (x)**.

---

### üîπ Simple vs. Multiple Linear Regression

- **Simple Linear Regression:** only **one** independent variable (1 feature).
- **Multiple Linear Regression:** **more than one** independent variable (multiple features).

A key requirement is that the **dependent variable (y) must be continuous**.

---

## ‚≠ê Geometric Interpretation

If the input data has **k dimensions** (k features or inputs), linear regression fits a **hyperplane** in a space with **k + 1 dimensions**.

- For **simple linear regression (k = 1)**, the model becomes a **line in 2D**.
- For **multiple regression (k > 1)**, the model becomes a **plane** or a **hyperplane** in higher dimensions.


---
## ‚≠ê Simple Linear Regression

**Simple linear regression** measures the relationship between an **independent variable** on a **dependent variable** and measures the correlation between them. 

In this method, by fitting a **linear equation** to the observed data, the relationship between two variables (independent and dependent) is modeled. Data related to regression problems are generally represented as a set of **input/output data pairs**.

# ‚≠ê Simple Linear Regression 
Simple linear regression models the relationship between **one independent variable (x)** and **one dependent variable (y)**.  
Its purpose is to:

- Measure how strongly x and y are related (correlation)
- Fit a straight line that predicts **y** from **x**

---

## üìå Regression Data Format

Data for regression is represented as **pairs of input/output values**:

$
(x_1, y_1), (x_2, y_2), \dots, (x_n, y_n)
$

Where:

- $x_i$: i-th input (scalar)
- $y_i$: i-th output (scalar)

Since this is **simple** regression, both are one-dimensional.

---

## ‚≠ê Goal of Simple Linear Regression

We want to find the **best-fitting straight line** that describes the relationship between x and y:


$y = w x + b$


Where:

- **w** ‚Üí slope (weight)
- **b** ‚Üí intercept (bias)
- **$y_pred$** ‚Üí predicted y value

---

## ‚≠ê Prediction Formula

For an input value \(x_i\), the model predicts:


$\hat{y_i} = w x_i + b$


This is the hypothesis (model output).

---

## ‚≠ê Loss Function (Mean Squared Error - MSE)

To measure how good our predictions are, we compare:

- predicted values $\hat{y_i}$
- actual values $y_i$

The most common loss function for linear regression is **MSE**:


$\text{MSE} = \frac{1}{n} \sum_{i=1}^n (\hat{y_i} - y_i)^2$


---

## ‚≠ê Substituting the Prediction Formula

Since:


$\hat{y_i} = w x_i + b$


The MSE becomes:


$\text{MSE} = \frac{1}{n} \sum_{i=1}^n (w x_i + b - y_i)^2
$

This function is minimized to find the best **w** and **b**.

---

## ‚≠ê Summary

| Concept | Meaning |
|--------|---------|
| Data | (x, y) pairs |
| Model | $y = wx + b$ |
| Prediction | $\hat{y} = wx + b$ |
| Loss | Mean Squared Error (MSE) |
| Goal | Find w and b that minimize error |
